```python
import os
import numpy as np
import cv2
import h5py
import random

def cv_imread( file_path  ):
    #print ( file_path)
    cv_img=cv2.imdecode(np.fromfile(file_path,dtype=np.uint8),-1)  
## imdecode读取的是rgb，如果后续需要opencv处理的话，需要转换成bgr，转换后图片颜色会变化  
    #cv_img=cv2.cvtColor(cv_img,cv2.COLOR_RGB2BGR)  
    return cv_img  
def rotate_bound1(image, angle):
    '''
     . 旋转图片
     . @param image    opencv读取后的图像
     . @param angle    (逆)旋转角度
    '''

    # img = cv2.imread("img/1.jpg")
    (h, w) = image.shape[:2]  # 返回(高,宽,色彩通道数),此处取前两个值返回
    # 抓取旋转矩阵(应用角度的负值顺时针旋转)。参数1为旋转中心点;参数2为旋转角度,正的值表示逆时针旋转;参数3为各向同性的比例因子
    M = cv2.getRotationMatrix2D((w / 2, h / 2), -angle, 1.0)
    # 计算图像的新边界维数
    newW = int((h * np.abs(M[0, 1])) + (w * np.abs(M[0, 0])))
    newH = int((h * np.abs(M[0, 0])) + (w * np.abs(M[0, 1])))
    # 调整旋转矩阵以考虑平移
    M[0, 2] += (newW - w) / 2
    M[1, 2] += (newH - h) / 2
    # 执行实际的旋转并返回图像
    return cv2.warpAffine(image, M, (newW, newH)) # borderValue 缺省，默认是黑色

def sub_image( img,rest2,rest3,rest4,w,length,width,interval,imglr_list,imghr_list,cnt):
    num = 0
    i = 0
    while ( i + w ) < width:
        j =0 
        while (j +w ) < length:
            sub_img1 = img[i:i+w,j:j+w,:]
            imghr_list.append(sub_img1) 
            sub_img2 = rest2 [i:i+w,j:j+w,:]
            imglr_list.append(sub_img2) 
      
            cnt = cnt +1
      
            imghr_list.append(sub_img1) 
            sub_img2 = rest3 [i:i+w,j:j+w,:]
            imglr_list.append(sub_img2) 
            
            cnt = cnt +1
     
            imghr_list.append(sub_img1) 
            sub_img2 = rest4 [i:i+w,j:j+w,:]
            imglr_list.append(sub_img2) 
     
            cnt = cnt + 1
            j = j+interval
            
        i = i+interval
    
    
    return imglr_list,imghr_list,cnt

def save_image_to_h5py(path,w):
    imglr_list=[]  #初始化
    imghr_list=[]  
    cn = 0
    cnt =0
    for dir_image in os.listdir(path):
        #读取文件
        img=cv_imread(os.path.join(path,dir_image))
        t = random.choice ( [0,1,2,3])
        if  t==1 :
            img = rotate_bound1 ( img,90)
        
        if t==2:
            img = rotate_bound1 ( img,180)
    
        if t == 3:
            img = rotate_bound1 ( img,270)
            
        width ,length, channels = img.shape 
       
        #print ( img.shape)
        rest = cv2.resize ( img,None, fx=0.5,fy =0.5)  # 下采样 
        #printnt ( rest.shape)
        rest2 = cv2.resize(rest,None, fx=2,fy =2,interpolation = cv2.INTER_CUBIC) # 上采样双三次插值
        w2,l2, c2 = rest2.shape
        width = np.min([w2,width])
        length = np.min([l2,length])
    
        rest = cv2.resize ( img,None, fx=0.33,fy =0.33)  # 下采样 
        #print ( rest.shape)
        rest3 = cv2.resize(rest,None, fx=3,fy =3,interpolation = cv2.INTER_CUBIC) # 上采样双三次插值
        w2, l2, c2 = rest3.shape
        width = np.min([w2,width])
        length = np.min([l2,length])
   
        rest = cv2.resize ( img,None, fx=0.25,fy =0.25)  # 下采样 
        #print ( rest.shape)
        rest4 = cv2.resize(rest,None, fx=4,fy =4,interpolation = cv2.INTER_CUBIC) # 上采样双三次插值
        w2, l2, c2 = rest4.shape
        width = np.min([w2,width])
        length = np.min([l2,length])
        print ( rest2.shape)
        
        imglr_list,imghr_list,cnt = sub_image( img,rest2,rest3,rest4,w,length,width,27,imglr_list,imghr_list,cnt)
        #imglr_list,imghr_list,cnt = sub_image( img,rest2,rest3,rest4,w,length,width,37,imglr_list,imghr_list,cnt)
        #imglr_list,imghr_list,cnt = sub_image( img,rest2,rest3,rest4,w,length,width,17,imglr_list,imghr_list,cnt)
        cn = cn+1
        print (cn)

    imglr_np = np.array(imglr_list)  #转为numpy的ndarray类型

    f = h5py.File(r'C:\e\SR_datasets\level6_process\vdsr\lr_vdsr.h5', 'w')  #写入文件
    f['image'] = imglr_np  #名称为image
    f.close()  #关闭文件
    
    imghr_np = np.array(imghr_list)  #转为numpy的ndarray类型

    f1 = h5py.File(r'C:\e\SR_datasets\level6_process\vdsr\hr_vdsr.h5', 'w')  #写入文件
    f1['image'] = imghr_np  #名称为image
    f1.close()  #关闭文件
    
    return cnt
#C:\e\SR_datasets\level6
filepath = r'C:\e\SR_datasets\level6'
num1 = save_image_to_h5py(filepath,41)
print ( num1)
```
