```python
import os
import numpy as np
import cv2
import h5py
import random

def cv_imread( file_path  ):
    #print ( file_path)
    cv_img=cv2.imdecode(np.fromfile(file_path,dtype=np.uint8),-1)  
## imdecode读取的是rgb，如果后续需要opencv处理的话，需要转换成bgr，转换后图片颜色会变化  
    #cv_img=cv2.cvtColor(cv_img,cv2.COLOR_RGB2BGR)  
    return cv_img  
def rotate_bound1(image, angle):
    '''
     . 旋转图片
     . @param image    opencv读取后的图像
     . @param angle    (逆)旋转角度
    '''

    # img = cv2.imread("img/1.jpg")
    (h, w) = image.shape[:2]  # 返回(高,宽,色彩通道数),此处取前两个值返回
    # 抓取旋转矩阵(应用角度的负值顺时针旋转)。参数1为旋转中心点;参数2为旋转角度,正的值表示逆时针旋转;参数3为各向同性的比例因子
    M = cv2.getRotationMatrix2D((w / 2, h / 2), -angle, 1.0)
    # 计算图像的新边界维数
    newW = int((h * np.abs(M[0, 1])) + (w * np.abs(M[0, 0])))
    newH = int((h * np.abs(M[0, 0])) + (w * np.abs(M[0, 1])))
    # 调整旋转矩阵以考虑平移
    M[0, 2] += (newW - w) / 2
    M[1, 2] += (newH - h) / 2
    # 执行实际的旋转并返回图像
    return cv2.warpAffine(image, M, (newW, newH)) # borderValue 缺省，默认是黑色

def sub_image( img,rest2,w,length,width,interval,imglr_list,imghr_list,cnt):
    num = 0
    i = 0
    while ( i + w ) < width:
        j =0 
        while (j +w ) < length:
            print (j+w)
           
            m1 = img[0:w,0:w*2,:]
            m1[0:w,w:w*2,:] = img[i:i+w,j:j+w,:]
            sub_img1 = img[i:i+w,j:j+w,:]
            #cv2.imwrite ("D:/xiao mai/n_l1l3/s_1_3pix2pix/l3/"+str(cnt)+".png",sub_img1) 
            #cv2.imshow('img',sub_img1)
            #cv2.waitKey(5000)
            #D:\xiao mai\n_l1l3\s_1_3\l1
            cv2.imwrite ("D:/xiao mai/pytorch-CycleGAN-and-pix2pix-master/pytorch-CycleGAN-and-pix2pix-master/datasets/pl1_l3/testB/"+str(cnt)+".png",sub_img1)
           # imghr_list.append(sub_img1) 
            print (j)
            sub_img2 = rest2 [i:i+w,j:j+w,:]
            m1[0:w,0:w,:] = rest2 [i:i+w,j:j+w,:]
           # imglr_list.append(sub_img2) 
            #cv2.imwrite ("D:/xiao mai/n_l1l3/s_1_3pix2pix/l1/"+str(cnt)+".png",sub_img2)
            cv2.imwrite ("D:/xiao mai/pytorch-CycleGAN-and-pix2pix-master/pytorch-CycleGAN-and-pix2pix-master/datasets/pl1_l3/testA/"+str(cnt)+".png",sub_img2)
            cv2.imwrite ("D:/xiao mai/pytorch-CycleGAN-and-pix2pix-master/pytorch-CycleGAN-and-pix2pix-master/datasets/pl1_l3c/test_s2/test/"+str(cnt)+".png",m1)
           
            cnt = cnt +1
           
        
            j = j+interval
            
        i = i+interval
    
    
    return imglr_list,imghr_list,cnt

def save_image_to_h5py(path1,path2,w,nu):
    imglr_list=[]  #初始化
    imghr_list=[]  
    cn = 0
    cnt =nu
    i=0
    while i <1:
        #读取文件
        #img=cv_imread(os.path.join(path,dir_image))
        
        himg=cv_imread(path1)
        limg = cv_imread(path2)    
        himg = himg[524:653,110:727,:]
        limg = limg[524:653,110:727,:]
        t = random.choice ( [0,1,2,3])
        t=0
        if  t==1 :
            himg = rotate_bound1 ( himg,90)
            limg = rotate_bound1 ( limg,90)
        
        if t==2:
            himg = rotate_bound1 ( himg,180)
            limg = rotate_bound1 ( limg,180)
    
        if t == 3:
            himg = rotate_bound1 ( himg,270)
            limg = rotate_bound1 ( limg,270)
        
    
    
    #img = img[0:480,0:320,:]
   # width ,length, channels = img.shape 
        width ,length, channels = himg.shape 

        imglr_list,imghr_list,cnt = sub_image( himg,limg,w,length,width,128,imglr_list,imghr_list,cnt)
        #imglr_list,imghr_list,cnt = sub_image( himg,limg,w,length,width,17,imglr_list,imghr_list,cnt)
        #imglr_list,imghr_list,cnt = sub_image( himg,limg,w,length,width,25,imglr_list,imghr_list,cnt)
        #imglr_list,imghr_list,cnt = sub_image( img,rest2,rest3,rest4,w,length,width,37,imglr_list,imghr_list,cnt)
        #imglr_list,imghr_list,cnt = sub_image( img,rest2,rest3,rest4,w,length,width,17,imglr_list,imghr_list,cnt)
        cn = cn+1
        i=i+1
        print (cn)
    
    #imglr_np = np.array(imglr_list)  #转为numpy的ndarray类型
    #f = h5py.File(r'D:\xiao mai\SR_datasets\level6_process\vdsr\lr_4.h5', 'w')  #写入文件
    #f['image'] = imglr_np  #名称为image
    #f.close()  #关闭文件
    
        
    #imghr_np = np.array(imghr_list)  #转为numpy的ndarray类型

    #f1 = h5py.File(r'D:\xiao mai\SR_datasets\level6_process\vdsr\hr_6.h5', 'w')  #写入文件
    #f1['image'] = imghr_np  #名称为image
    #f1.close()  #关闭文件
    
    return cnt


c1=0
c3=0
num1=0

while c3<319:
    filepath = "D:/xiao mai/n_l1l3/nl1_image/"+str(c1)+".png"
    filepath2 = "D:/xiao mai/n_l1l3/new_l3/"+str(c3)+".png" #D:\xiao mai\n_l1l3\l3image
    #num1 = save_image_to_h5py(filepath2,filepath,64,num1)
   
    #c3=c3+1
    #filepath2 = "D:/xiao mai/n_l1l3/l3image/"+str(c3)+".png" #D:\xiao mai\n_l1l3\l3image
    #num1 = save_image_to_h5py(filepath2,filepath,128,num1)
    c3=c3+1
    c1=c1+1
    
   
#C:\e\SR_datasets\level6
filepath = r'D:\xiao mai\l1_img\84_7.png'
filepath2 = r'D:\xiao mai\l3_img\47220_7.png'

#num1 = save_image_to_h5py(filepath2,filepath,64)
print ( num1)
```
