```python
import cv2
import numpy as np
import glob
import random

def cv_imread( file_path ="C:/e/学习资料/2/毕业设计/data/picture - level4/0.jpeg" ):
    cv_img=cv2.imdecode(np.fromfile(file_path,dtype=np.uint8),-1)  
## imdecode读取的是rgb，如果后续需要opencv处理的话，需要转换成bgr，转换后图片颜色会变化  
    #cv_img=cv2.cvtColor(cv_img,cv2.COLOR_RGB2BGR)  
    return cv_img  

def rotate_bound1(image, angle):
    '''
     . 旋转图片
     . @param image    opencv读取后的图像
     . @param angle    (逆)旋转角度
    '''

    # img = cv2.imread("img/1.jpg")
    (h, w) = image.shape[:2]  # 返回(高,宽,色彩通道数),此处取前两个值返回
    # 抓取旋转矩阵(应用角度的负值顺时针旋转)。参数1为旋转中心点;参数2为旋转角度,正的值表示逆时针旋转;参数3为各向同性的比例因子
    M = cv2.getRotationMatrix2D((w / 2, h / 2), -angle, 1.0)
    # 计算图像的新边界维数
    newW = int((h * np.abs(M[0, 1])) + (w * np.abs(M[0, 0])))
    newH = int((h * np.abs(M[0, 0])) + (w * np.abs(M[0, 1])))
    # 调整旋转矩阵以考虑平移
    M[0, 2] += (newW - w) / 2
    M[1, 2] += (newH - h) / 2
    # 执行实际的旋转并返回图像
    return cv2.warpAffine(image, M, (newW, newH)) # borderValue 缺省，默认是黑色


def process ( file_path,num,w,interval):
    img = cv_imread ( file_path )
    width ,length, channels = img.shape 
    t = random.choice ( [0,1,2,3])
    if  t==1 :
        img = rotate_bound1 ( img,90)
        
    if t==2:
        img = rotate_bound1 ( img,180)
    
    if t == 3:
        img = rotate_bound1 ( img,270)
        
    
    
    #img = img[0:480,0:320,:]
    width ,length, channels = img.shape 
    #print ( img.shape)
    rest = cv2.resize ( img,None, fx=0.5,fy =0.5)  # 下采样 
    #print ( rest.shape)
    rest2 = cv2.resize(rest,None, fx=2,fy =2,interpolation = cv2.INTER_CUBIC) # 上采样双三次插值
    w2, l2, c2 = rest2.shape
    width = np.min([w2,width])
    length = np.min([l2,length])
    
    rest = cv2.resize ( img,None, fx=0.33,fy =0.33)  # 下采样 
    #print ( rest.shape)
    rest3 = cv2.resize(rest,None, fx=3,fy =3,interpolation = cv2.INTER_CUBIC) # 上采样双三次插值
    w2, l2, c2 = rest3.shape
    width = np.min([w2,width])
    length = np.min([l2,length])
   
    rest = cv2.resize ( img,None, fx=0.25,fy =0.25)  # 下采样 
    #print ( rest.shape)
    rest4 = cv2.resize(rest,None, fx=4,fy =4,interpolation = cv2.INTER_CUBIC) # 上采样双三次插值
    w2, l2, c2 = rest4.shape
    width = np.min([w2,width])
    length = np.min([l2,length])
    print ( rest2.shape)
    
    i = 0
    cnt = num
    
    while ( i + w ) < width:
        j =0 
        while (j +w ) < length:
            sub_img1 = img[i:i+41,j:j+41,:]
            re = cv2.imwrite ("C:/e/SR_datasets/SR/processed/HR/"+str(cnt)+".jpg",sub_img1)
            sub_img2 = rest2 [i:i+41,j:j+41,:]
            re2 =cv2.imwrite ("C:/e/SR_datasets/SR/processed/LR/"+str(cnt)+".jpg",sub_img2)
            
            cnt = cnt +1
            sub_img1 = img[i:i+41,j:j+41,:]
            re = cv2.imwrite ("C:/e/SR_datasets/SR/processed/HR/"+str(cnt)+".jpg",sub_img1)
            sub_img2 = rest3 [i:i+41,j:j+41,:]
            re2 =cv2.imwrite ("C:/e/SR_datasets/SR/processed/LR/"+str(cnt)+".jpg",sub_img2)
            
            cnt = cnt +1
            sub_img1 = img[i:i+41,j:j+41,:]
            re = cv2.imwrite ("C:/e/SR_datasets/SR/processed/HR/"+str(cnt)+".jpg",sub_img1)
            sub_img2 = rest4 [i:i+41,j:j+41,:]
            re2 =cv2.imwrite ("C:/e/SR_datasets/SR/processed/LR/"+str(cnt)+".jpg",sub_img2)
            
            cnt = cnt + 1
            j = j+interval
            
            
        i = i+interval
    
    return cnt
            
 
def processTest ( file_path,num,w1):
    img = cv_imread ( file_path )
    width ,length, channels = img.shape 
    if  width<length :
        img = rotate_bound1 ( img,90)
    
   # img = img[0:480,0:320,:]
    width ,length, channels = img.shape 
    #print ( img.shape)
    rest = cv2.resize ( img,None, fx=0.5,fy =0.5)  # 下采样 
    #print ( rest.shape)
    rest2 = cv2.resize(rest,None, fx=2,fy =2,interpolation = cv2.INTER_CUBIC) # 上采样双三次插值
    print ( rest2.shape)
    #re2 =cv2.imwrite ("C:/e/SR_datasets/SR/test/butterfly/"+str(cnt)+"l.jpg",sub_img2)
    i = 0
    cnt = num
    
    while ( i + w1 ) < width:
        
        j =0 
        while (j +w1 ) < length:
            sub_img1 = img[i:i+w,j:j+w,:]
            re = cv2.imwrite ("C:/e/SR_datasets/SR/test/butterfly/"+str(cnt)+"h.jpg",sub_img1)
            sub_img2 = rest2 [i:i+w,j:j+w,:]
            re2 =cv2.imwrite ("C:/e/SR_datasets/SR/test/butterfly/"+str(cnt)+"l.jpg",sub_img2)
            j = j+w1
            cnt = cnt +1
            
        i = i+w1
    
   
    
    return cnt
            
def process_batch (filepath,n1,w,interval ):
    num = n1
    for jpgfile in glob.glob(filepath):
        num2 = process(jpgfile,num,w,interval)
        num = num2
        
    return num   
    
'''
  cv2.namedWindow("image")
    cv2.imshow("image", img)    # 显示图像
    cv2.waitKey(5000)   # 默认为0，无限等待
    cv2.destroyAllWindows()  
'''

file = 'C:/e/SR_datasets/archive2/train/12074.jpg' # 25098
filepath = r'C:\e\SR_datasets\SR\SR_training_datasets\BSDS200\*.png'
t_num = 0
#t_num = process_batch(filepath = filepath,n1 = 0,w = 96,interval= 41)
ct = t_num
#t_num = process_batch(filepath = filepath,n1=ct,w=96,interval=13)
ct = t_num
#t_num = process_batch(filepath = filepath,n1=ct,w=64,interval=19)
ct = t_num
#t_num = process_batch(filepath = filepath,n1=ct,w=64,interval=27)
print ( ct)

#print ('一共 ',t_num )
#li = glob.glob(r'C:\e\SR_datasets\SR\SR_training_datasets\BSDS200\*.png')
#print (len (li))
#print ( li[0])

test_img = 'C:/e/SR_datasets/SR/SR_testing_datasets/Set5/butterfly.png'
n1 = 0#processTest( test_img,0,64)
print (n1)
```
